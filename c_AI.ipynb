```python
from b_Game import Game
from a_mastermindGameLogic import MastermindLogic as logic
import pandas as pd
import random as r
import ast  # Import the ast module


# Assuming the CSV file contains a single column
possibleGuesses = pd.read_csv('solutions.csv', header=None, squeeze=True)

# Convert string representations to lists
possibleGuesses = possibleGuesses.apply(ast.literal_eval)
```

```python
guess_pin_pairing = []

for row in possibleGuesses:
    guess_pin_pairing.append((row,None))


# Counters
wins = 0
losses = 0
guesses = 0


for i in range(10000):
    currentGame = Game()

    guess_pin_pairing = []

    for row in possibleGuesses:
        guess_pin_pairing.append((row,None))

    while True:
        currentGuess = r.choice(guess_pin_pairing)[0]

        currentGame.guessCode(currentGuess)
        if currentGame.gameOver: break

        for i in range(len(guess_pin_pairing)):
            guess_pin_pairing[i] = (guess_pin_pairing[i][0], logic.placePins(currentGuess, guess_pin_pairing[i][0]))

        guess_pin_pairing = [x for x in guess_pin_pairing 
                            if 
                            (x[1][0] == currentGame.pins[0]) ]
                            # or 
                        #  (x[1][0] == currentGame.pins[0] and x[1][1] <= currentGame.pins[1])
                        # ]
    if currentGame.correct:
        wins += 1
        guesses += len(currentGame.guesses)
    else: losses += 1

print("Wins: ", wins)
print("Losses: ", losses)
print("Average guesses to win: ", guesses/wins)

```
